1. Who is Bob’s best customer by total sales?

```python
# best customer by Total Sales
qn1 = df.groupby("Customer")["Total Sales"].sum().sort_values(ascending=False)
qn1.plot(
    kind="bar",
    figsize=(10, 5),
    title="Best Customer by Total Sales",
    ylabel="Total Sales",
    xlabel="Customer",
    grid=True,
)
qq = qn1.to_frame()
qq
```

```python
# worst customer by Total Sales
qn2 = df.groupby("Customer")["Total Sales"].sum().sort_values(ascending=True)
qn2.plot(
    kind="bar",
    figsize=(10, 5),
    title="Worst Customer by Total Sales",
    ylabel="Total Sales",
    xlabel="Customer",
    grid=True,
)
```
